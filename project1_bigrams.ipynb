# Project1:Bigrams

**Find bigrams in the attached text. Bigrams are word pairs and their counts. To build them do the following:**

```python
from collections import Counter
""" Here we are reading the entire content of the file as a single string and
    converting them to lowercase to make the analysis case-insensitive.
"""
with open('nyt.txt', 'r') as file:
    text = file.read().lower()
text
```

**1.Tokenize by word.**

```python
#Spliting the text into a list of words, where each word is a token.
words = text.split()
words
```

**2.Create two almost-duplicate files of words, off by one line, using tail.**

```python
# Createing two lists of words, shifted by one
words1 = words[:-1]  # All words except the last one
words2 = words[1:]   # All words except the first one
```

**3.Paste them together so as to get word(i) and word(i +1) on the same line.**


```python
# Form bigrams by zipping the two lists
bigrams = list(zip(words1, words2))
# Count bigram occurrences
bigram_counts = Counter(bigrams)
bigram_counts
```

**4.Then, after you have the data from the procedure above: Provide the commands to find the 10 most common bigrams.**

```python
#Getting the 10 most common bigrams
top_10_bigrams = bigram_counts.most_common(10)
top_10_bigrams
```

```python
# Displaying the output
for bigram, count in top_10_bigrams:
    print(f'Bigram: {bigram}, Count: {count}')
```
